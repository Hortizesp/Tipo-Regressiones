```python
import numpy as np
```

```python
class LinearRegression:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.b1, self.b0 = self.regresion_lineal()

    def regresion_lineal(self):
        n = len(self.x)
        mean_x = np.mean(self.x)
        mean_y = np.mean(self.y)

        numerator = np.sum((self.x - mean_x) * (self.y - mean_y))
        denominator = np.sum((self.x - mean_x) ** 2)

        b1 = numerator / denominator
        b0 = mean_y - (b1 * mean_x)

        return b1, b0

    def calcular_metricas(self):
        y_pred = [self.b0 + self.b1 * xi for xi in self.x]
        mean_y = np.mean(self.y)
            # R2
        ssr = np.sum((y_pred - mean_y) ** 2)
        sst = np.sum((self.y - mean_y) ** 2)
        r_squared = ssr / sst
            # MSE
        mse = np.mean((self.y - y_pred) ** 2)
            # RMSE
        rmse = np.sqrt(mse)

        return r_squared, mse, rmse
```

```python
regression= LinearRegression(x,y)

b1 = regression.b1
b0 = regression.b0

print("Coeficiente de la pendiente:", b1)
#print("Coeficiente del intercepto:", b0)

# Calcular las métricas utilizando el método calcular_metricas
r_squared, mse, rmse = regression.calcular_metricas()
print("R2:", r_squared)
print("MSE:", mse)
print("RMSE:", rmse)
```
